```python
from pymongo import MongoClient
import os
from dotenv import load_dotenv
load_dotenv()
COSMOS_MONGO_URI = os.getenv('COSMOS_MONGO_URI')


# Create the client
client = MongoClient(COSMOS_MONGO_URI)
# List the databases
try:
    databases = client.list_database_names()
    print("✅ Connection successful! Databases:", databases)
except Exception as e:
    print("❌ Connection failed:", e)
```

```python
db = client["quickstartDB"]

db.list_collection_names()
```

```python
collection = db['sampleCollection']

for document in collection.find():
    print(document)
```

```python
import pandas as pd
news = pd.read_csv("news_database.csv")
news.head()
```

```python
news_dict = news.to_dict(orient='records')
news_dict
```

```python
# create news collection
collection = db['news']

collection.insert_many(news_dict)

print(f"✅ Inserted {len(news_dict)} documents into the 'news' collection.")
```

```python
for document in collection.find():
    print(document)
```
